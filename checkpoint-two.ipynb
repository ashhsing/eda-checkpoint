After analyzing data, the family you are working with decides to narrow down their choices to breeds that rank at a 4 or 5 for being good with children. 

```python
# Drop rows in which dog breeds >=4 in "good_with_children" stay
dogs.drop(dogs[dogs['good_with_children'] < 4].index, inplace = True)
```

```python
dogs.plot(x='Name', y='good_with_children', style='o', color='c')
```

```python
dogs.drop(dogs[dogs['trainability'] < 4].index, inplace = True)
```

```python
dogs.plot(x='Name', y='trainability', style='o', color='khaki')
```

```python
dogs.drop(dogs[dogs['good_with_other_dogs'] < 4].index, inplace = True)
```
